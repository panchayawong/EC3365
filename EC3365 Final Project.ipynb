```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Import data sets and clean data
emissions = pd.read_csv("https://raw.githubusercontent.com/pforwork/EC3365/main/emissions.csv", delimiter=',', skiprows=1)
emissions
```

```python
energy = pd.read_csv("https://raw.githubusercontent.com/pforwork/EC3365/main/energyconsumption.csv", delimiter=',', skiprows=1)
energy.fillna(energy.median(), inplace=True)
energy.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
energy.dropna(subset=['API'], inplace=True)
energy.dropna(how='all', inplace=True)

for col in energy.columns[2:]:
    if energy[col].dtype in ['float64', 'int64']:
        energy[col].fillna(energy[col].median(), inplace=True)
energy
```

```python
countries_df = gpd.read_file("https://raw.githubusercontent.com/pforwork/EC3365/main/world-countries.json")
countries_df
```

```python
pwt = pd.read_csv("https://raw.githubusercontent.com/pforwork/EC3365/main/pwt1001.csv")
pwt = pwt[pwt['year'] >= 1980]   
pwt['emp'] = pwt['emp'].astype(float)
pwt.drop(columns=['countrycode', 'currency_unit'], inplace=True)
pwt.sort_values(by=['country', 'year'], inplace=True)
pwt.fillna(method='ffill', inplace=True)
pwt.fillna(method='bfill', inplace=True)
pwt = pwt.dropna().reset_index(drop=True)
pwt
```
